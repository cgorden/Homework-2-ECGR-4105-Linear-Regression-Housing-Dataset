```python
#HW #2: Housing Price Prediction using Linear Regression with Gradient Descent

#mount google drive
from google.colab import drive
drive.mount('/content/drive')

#libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
import seaborn as sns

#rand Seed
np.random.seed(42)

#TA's please update file path below with location of CSV File to run on one's own VM
FilePath = '/content/drive/MyDrive/Year 4: 2025/Machine Learning/Homeworks/HW 2/Housing.csv'
DataFile = pd.read_csv(FilePath)

print("Dataset shape:", DataFile.shape)
print("\nDataset info:")
print(DataFile.info())
print("\nFirst few rows:")
print(DataFile.head())

print("\nMissing values:")
print(DataFile.isnull().sum())

print("Dataset shape:", DataFile.shape)
print("\nDataset info:")
print(DataFile.info())
print("\nFirst few rows:")
print(DataFile.head())

print("\nMissing values:")
print(DataFile.isnull().sum())

#converts csv variables into boolean for processing
CatColumns = ['mainroad', 'guestroom', 'basement', 'hotwaterheating',
              'airconditioning', 'prefarea']

#checks if columns exist
for ColName in CatColumns:
    if ColName in DataFile.columns:
        DataFile[ColName] = DataFile[ColName].map({'yes': 1, 'no': 0})

#enumerates data
if 'furnishingstatus' in DataFile.columns:
    StatsMap = {'furnished': 2, 'semi-furnished': 1, 'unfurnished': 0}
    DataFile['furnishingstatus'] = DataFile['furnishingstatus'].map(StatsMap)

print("\nDataset after preprocessing:")
print(DataFile.head())

class LinearRegression:

    def __init__(self, LearnRate=0.01, MaxIterations=1000, Reg=None, Lambda=0.01):
        self.LearnRate = LearnRate
        self.MaxIterations = MaxIterations
        self.Reg = Reg
        self.Lambda = Lambda
        self.Weights = None
        self.Bias = None
        self.TrainLoss = []
        self.ValLoss = []

        #trains regression model
    def fit(self, XTrain, YTrain, XVal=None, YVal=None):
        nFeat = XTrain.shape[1]
        self.Weights = np.zeros(nFeat)
        self.Bias = 0

        #Training loop
        for Iteration in range(self.MaxIterations):
            #passes Forward
            PredTrain = self.predict(XTrain)

            #Finds training loss and records
            TrainLossVal = self.ComputeLoss(YTrain, PredTrain)
            self.TrainLoss.append(TrainLossVal)

            #Finds validation loss
            if XVal is not None and YVal is not None:
                PredVal = self.predict(XVal)
                ValLossVal = self.ComputeLoss(YVal, PredVal, IsTrain=False)
                self.ValLoss.append(ValLossVal)

            #Finds the gradient
            GradW, GradB = self.CompGradient(XTrain, YTrain, PredTrain)

            #updates values from computations
            self.Weights -= self.LearnRate * GradW
            self.Bias -= self.LearnRate * GradB

            if Iteration % 100 == 0:
                print(f"Iteration {Iteration}, Train Loss: {TrainLossVal:.4f}", end="")
                if XVal is not None:
                    print(f", Val Loss: {ValLossVal:.4f}")
                else:
                    print()

    #predicts using slope formula (y=mx+b)
    def predict(self, XData):
        return np.dot(XData, self.Weights) + self.Bias

    def ComputeLoss(self, YTrue, YPred, IsTrain=True):
        MseLoss = np.mean((YTrue - YPred) ** 2)

        if IsTrain and self.Reg == 'ridge':
            RegTerm = self.Lambda * np.sum(self.Weights ** 2)
            return MseLoss + RegTerm

        return MseLoss

    def CompGradient(self, XData, YTrue, YPred):
        #Compute gradients for weights and bia
        NumSamples = XData.shape[0]

        #finds gradients
        GradW = -(2/NumSamples) * np.dot(XData.T, (YTrue - YPred))
        GradB = -(2/NumSamples) * np.sum(YTrue - YPred)

        #places regularization gradient
        if self.Reg == 'ridge':
            GradW += 2 * self.Lambda * self.Weights

        return GradW, GradB

    def score(self, XData, YData):
        #R^2 score
        YPred = self.predict(XData)
        SsRes = np.sum((YData - YPred) ** 2)
        SsTot = np.sum((YData - np.mean(YData)) ** 2)
        return 1 - (SsRes / SsTot)

def PlotLoss(ModelDict, PlotTitle):
    plt.figure(figsize=(12, 8))

    ColorList = ['blue', 'red', 'green', 'orange', 'purple']

    for Index, (ModelName, ModelObj) in enumerate(ModelDict.items()):
        PlotColor = ColorList[Index % len(ColorList)]
        plt.plot(ModelObj.TrainLoss, color=PlotColor, linestyle='-',
                label=f'{ModelName} - Training', alpha=0.7)
        if ModelObj.ValLoss:
            plt.plot(ModelObj.ValLoss, color=PlotColor, linestyle='--',
                    label=f'{ModelName} - Validation', alpha=0.7)

    plt.xlabel('Iterations')
    plt.ylabel('Loss (MSE)')
    plt.title(PlotTitle)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

def EvaluateModel(ModelObj, XTrain, YTrain, XTest, YTest, ModelName):
    TrainScore = ModelObj.score(XTrain, YTrain)
    TestScore = ModelObj.score(XTest, YTest)

    print(f"\n{ModelName} Results:")
    print(f"Training R²: {TrainScore:.4f}")
    print(f"Testing R²: {TestScore:.4f}")
    print(f"Final Training Loss: {ModelObj.TrainLoss[-1]:.4f}")
    if ModelObj.ValLoss:
        print(f"Final Validation Loss: {ModelObj.ValLoss[-1]:.4f}")

    return TrainScore, TestScore

#===============================================================
#Problem 1A

print("\n" + "="*80)
print("PROBLEM 1A:")
print("Features: area, bedrooms, bathrooms, stories, parking")
print("="*80)

#5 features selected from csv
Features1a = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking']
XData1a = DataFile[Features1a].values
YTarget = DataFile['price'].values

#80/20% split from csv data
XTrain1a, XTest1a, YTrain, YTest = train_test_split(XData1a, YTarget, test_size=0.2, random_state=42)

print(f"Training set shape: {XTrain1a.shape}")
print(f"Test set shape: {XTest1a.shape}")

#arbitrary learning rates to find optimal value
LearnRates = [0.01, 0.03, 0.05, 0.1]
Models1a = {}

for LrVal in LearnRates:
    print(f"\nTraining with learning rate: {LrVal}")
    ModelObj = LinearRegression(LearnRate=LrVal, MaxIterations=1000)
    ModelObj.fit(XTrain1a, YTrain, XTest1a, YTest)
    Models1a[f'LR_{LrVal}'] = ModelObj
    EvaluateModel(ModelObj, XTrain1a, YTrain, XTest1a, YTest, f'Model LR={LrVal}')

# Print best parameters for Problem 1a
print("\nBest Parameters for Problem 1a (LR = 0.01):")
print("Weights:", BestModel1a.Weights)
print("Bias:", BestModel1a.Bias)

#1A Plot results
PlotLoss(Models1a, 'Problem 1a: Training and Validation Losses (Different Learning Rates)')

#===============================================================
#Problem 1B

print("\n" + "="*80)
print("PROBLEM 1b:")
print("Features: area, bedrooms, bathrooms, stories, mainroad, guestroom,")
print("         basement, hotwaterheating, airconditioning, parking, prefarea")
print("="*80)

#11 features selected from CSV
Features1b = ['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad',
              'guestroom', 'basement', 'hotwaterheating', 'airconditioning',
              'parking', 'prefarea']
XData1b = DataFile[Features1b].values

#80/20% split from csv data
XTrain1b, XTest1b, YTrain, YTest = train_test_split(XData1b, YTarget, test_size=0.2, random_state=42)

print(f"Training set shape: {XTrain1b.shape}")
print(f"Test set shape: {XTest1b.shape}")

# Train with different learning rates
Models1b = {}

for LrVal in LearnRates:
    print(f"\nTraining with learning rate: {LrVal}")
    ModelObj = LinearRegression(LearnRate=LrVal, MaxIterations=1000)
    ModelObj.fit(XTrain1b, YTrain, XTest1b, YTest)
    Models1b[f'LR_{LrVal}'] = ModelObj
    EvaluateModel(ModelObj, XTrain1b, YTrain, XTest1b, YTest, f'Model LR={LrVal}')

# Print best parameters for Problem 1b
print("\nBest Parameters for Problem 1b (LR = 0.01):")
print("Weights:", BestModel1b.Weights)
print("Bias:", BestModel1b.Bias)

# Plot results
PlotLoss(Models1b, 'Problem 1b: Training and Validation Losses (Different Learning Rates)')

# Compare linear regression models from 1a and 1b
print("\n" + "="*50)
print("COMPARISON: Problem 1a vs 1b")
print("="*50)

BestLr = 0.01 #adjust based on actual value
BestModel1a = Models1a[f'LR_{BestLr}']
BestModel1b = Models1b[f'LR_{BestLr}']

print(f"Problem 1a (5 features) - Test R²: {BestModel1a.score(XTest1a, YTest):.4f}")
print(f"Problem 1b (11 features) - Test R²: {BestModel1b.score(XTest1b, YTest):.4f}")

#===============================================================
#Problem 2A
#Normalization and Standarization w/ 5 Features

print("\n" + "="*80)
print("PROBLEM 2A: ")
print("="*80)

#Normization
ScalerNorm = MinMaxScaler()
XTrain1aNorm = ScalerNorm.fit_transform(XTrain1a)
XTest1aNorm = ScalerNorm.transform(XTest1a)

#standardization
ScalerStd = StandardScaler()
XTrain1aStd = ScalerStd.fit_transform(XTrain1a)
XTest1aStd = ScalerStd.transform(XTest1a)

Models2a = {}

#Train with Normalization
print("\nTraining with Normalization:")
ModelNorm = LinearRegression(LearnRate=0.01, MaxIterations=1000)
ModelNorm.fit(XTrain1aNorm, YTrain, XTest1aNorm, YTest)
Models2a['Normalization'] = ModelNorm
EvaluateModel(ModelNorm, XTrain1aNorm, YTrain, XTest1aNorm, YTest, 'Normalization')

#Train with standardization
print("\nTraining with standardization:")
ModelStd = LinearRegression(LearnRate=0.01, MaxIterations=1000)
ModelStd.fit(XTrain1aStd, YTrain, XTest1aStd, YTest)
Models2a['Standardization'] = ModelStd
EvaluateModel(ModelStd, XTrain1aStd, YTrain, XTest1aStd, YTest, 'Standardization')

#comparsion basline
Models2a['Baseline'] = BestModel1a

# Print best parameters for Problem 2a Normalization
print("\nBest Parameters for Problem 2a - Normalization:")
print("Weights:", ModelNorm.Weights)
print("Bias:", ModelNorm.Bias)

# Print best parameters for Problem 2a Standardization
print("\nBest Parameters for Problem 2a - Standardization:")
print("Weights:", ModelStd.Weights)
print("Bias:", ModelStd.Bias)

# Plot results for Problem 2a
PlotLoss(Models2a, 'Problem 2A: Normalization vs Standardization vs Baseline (5 features)')

#===============================================================
#Problem 2B
#Normalization and Standarization w/ 11 Features

print("\n" + "="*80)
print("PROBLEM 2B:")
print("="*80)

#Normalization
ScalerNorm1b = MinMaxScaler()
XTrain1bNorm = ScalerNorm1b.fit_transform(XTrain1b)
XTest1bNorm = ScalerNorm1b.transform(XTest1b)

#standardization
ScalerStd1b = StandardScaler()
XTrain1bStd = ScalerStd1b.fit_transform(XTrain1b)
XTest1bStd = ScalerStd1b.transform(XTest1b)

Models2b = {}

#Train with Normalization
print("\nTraining with Normalization:")
ModelNorm1b = LinearRegression(LearnRate=0.01, MaxIterations=1000)
ModelNorm1b.fit(XTrain1bNorm, YTrain, XTest1bNorm, YTest)
Models2b['Normalization'] = ModelNorm1b
EvaluateModel(ModelNorm1b, XTrain1bNorm, YTrain, XTest1bNorm, YTest, 'Normalization')

#Train with standardization
print("\nTraining with standardization:")
ModelStd1b = LinearRegression(LearnRate=0.01, MaxIterations=1000)
ModelStd1b.fit(XTrain1bStd, YTrain, XTest1bStd, YTest)
Models2b['Standardization'] = ModelStd1b
EvaluateModel(ModelStd1b, XTrain1bStd, YTrain, XTest1bStd, YTest, 'Standardization')

#comparsion basline
Models2b['Baseline'] = BestModel1b

# Print best parameters for Problem 2b Normalization
print("\nBest Parameters for Problem 2b - Normalization:")
print("Weights:", ModelNorm1b.Weights)
print("Bias:", ModelNorm1b.Bias)

# Print best parameters for Problem 2b Standardization
print("\nBest Parameters for Problem 2b - Standardization:")
print("Weights:", ModelStd1b.Weights)
print("Bias:", ModelStd1b.Bias)

# Plot results for Problem 2b
PlotLoss(Models2b, 'Problem 2b: Normalization vs Standardization vs Baseline (11 features)')

#===============================================================
#Problem 3A
#Ridge Regularization (5 features)

print("\n" + "="*80)
print("PROBLEM 3a: Ridge Regularization (5 features)")
print("="*80)

# Determine best scaling method from Problem 2a
BestScaling2a = 'Standardization'  # Adjust based on your results
if BestScaling2a == 'Normalization':
    XTrainBest = XTrain1aNorm
    XTestBest = XTest1aNorm
else:
    XTrainBest = XTrain1aStd
    XTestBest = XTest1aStd

#Different regularization parameters
LambdaVals = [0.001, 0.01, 0.1, 1.0]
Models3a = {}

for LamVal in LambdaVals:
    print(f"\nTraining with Ridge regularization (λ={LamVal}):")
    ModelRidge = LinearRegression(LearnRate=0.01, MaxIterations=1000,
                                 Reg='ridge', Lambda=LamVal)
    ModelRidge.fit(XTrainBest, YTrain, XTestBest, YTest)
    Models3a[f'Ridge_λ_{LamVal}'] = ModelRidge
    EvaluateModel(ModelRidge, XTrainBest, YTrain, XTestBest, YTest, f'Ridge λ={LamVal}')

# Add baseline for comparison
Models3a['Baseline_2a'] = Models2a[BestScaling2a]

# Print best parameters for Problem 3a (choose the lambda with best validation loss)
BestLam3a = min(LambdaVals, key=lambda lam: Models3a[f'Ridge_λ_{lam}'].ValLoss[-1])
BestModel3a = Models3a[f'Ridge_λ_{BestLam3a}']

print(f"\nBest Parameters for Problem 3a (Ridge λ={BestLam3a}):")
print("Weights:", BestModel3a.Weights)
print("Bias:", BestModel3a.Bias)

#Plot results
PlotLoss(Models3a, f'Problem 3a: Ridge Regularization with {BestScaling2a} (5 features)')

#===============================================================
#Problem 3B
#Ridge Regularization (11 features)

print("\n" + "="*80)
print("PROBLEM 3B:")
print("="*80)

# Determine best scaling method from Problem 2b
BestScaling2b = 'Standardization'  # Adjust based on your results
if BestScaling2b == 'Normalization':
    XTrainBest1b = XTrain1bNorm
    XTestBest1b = XTest1bNorm
else:
    XTrainBest1b = XTrain1bStd
    XTestBest1b = XTest1bStd

Models3b = {}

for LamVal in LambdaVals:
    print(f"\nTraining with Ridge regularization (λ={LamVal}):")
    ModelRidge = LinearRegression(LearnRate=0.01, MaxIterations=1000,
                                 Reg='ridge', Lambda=LamVal)
    ModelRidge.fit(XTrainBest1b, YTrain, XTestBest1b, YTest)
    Models3b[f'Ridge_λ_{LamVal}'] = ModelRidge
    EvaluateModel(ModelRidge, XTrainBest1b, YTrain, XTestBest1b, YTest, f'Ridge λ={LamVal}')

# Add baseline
Models3b['Baseline_2b'] = Models2b[BestScaling2b]

# Print best parameters for Problem 3b (choose the lambda with best validation loss)
BestLam3b = min(LambdaVals, key=lambda lam: Models3b[f'Ridge_λ_{lam}'].ValLoss[-1])
BestModel3b = Models3b[f'Ridge_λ_{BestLam3b}']

print(f"\nBest Parameters for Problem 3b (Ridge λ={BestLam3b}):")
print("Weights:", BestModel3b.Weights)
print("Bias:", BestModel3b.Bias)

# Plot results
PlotLoss(Models3b, f'Problem 3b: Ridge Regularization with {BestScaling2b} (11 features)')

#===============================================================
#Summary

print("\n" + "="*80)
print("Summary")
print("="*80)

SummaryResults = []

# Problem 1a
TestScore1a = BestModel1a.score(XTest1a, YTest)
SummaryResults.append(['Problem 1a (5 features)', 'Baseline', f'{TestScore1a:.4f}'])

# Problem 1b
TestScore1b = BestModel1b.score(XTest1b, YTest)
SummaryResults.append(['Problem 1b (11 features)', 'Baseline', f'{TestScore1b:.4f}'])

# Problem 2a
TestScore2aNorm = ModelNorm.score(XTest1aNorm, YTest)
TestScore2aStd = ModelStd.score(XTest1aStd, YTest)
SummaryResults.append(['Problem 2a (5 features)', 'Normalization', f'{TestScore2aNorm:.4f}'])
SummaryResults.append(['Problem 2a (5 features)', 'Standardization', f'{TestScore2aStd:.4f}'])

# Problem 2b
TestScore2bNorm = ModelNorm1b.score(XTest1bNorm, YTest)
TestScore2bStd = ModelStd1b.score(XTest1bStd, YTest)
SummaryResults.append(['Problem 2b (11 features)', 'Normalization', f'{TestScore2bNorm:.4f}'])
SummaryResults.append(['Problem 2b (11 features)', 'Standardization', f'{TestScore2bStd:.4f}'])

# Create DataFrame for better visualization
SummaryDf = pd.DataFrame(SummaryResults, columns=['Problem', 'Method', 'Test R²'])
print(SummaryDf)
```
